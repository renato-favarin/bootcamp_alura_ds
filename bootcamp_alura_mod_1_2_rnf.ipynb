# Desafio: Separar por região do país

```
# separar o indicador de reião presente no nome de cada estado
usaveis['Regiao'] = usaveis.index.str[0]

# criar um dicionário com as regiões do pais
regioes = {1:'N', 2:'NE', 3:'SE', 4:'S', 5:'CO'}


def busca_regiao(i):
    return regioes[int(i)]

# aplicar o código criado no dicionário para cada estado
usaveis['Regiao'] = usaveis['Regiao'].map(busca_regiao)

```

```
# Outra maneira é usar o lambda junto com o map:
# O lambda é uma função anônima do python, esse código é equivalente a:

# separar o indicador de reião presente no nome de cada estado
usaveis['Regiao'] = usaveis.index.str[0]

# criar um dicionário com as regiões do pais
regioes = {1:'N', 2:'NE', 3:'SE', 4:'S', 5:'CO'}

# aplicar o código criado no dicionário para cada estado
usaveis['Regiao'] = usaveis['Regiao'].map(lambda i: regioes[int(i)])

usaveis['Regiao']
```

```
#Plotando por região do Brasil

usaveis_regiao = usaveis.groupby(['Regiao']).sum()

ax = usaveis_regiao.T.plot(figsize=(12,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor total gasto nos procedimentos hospitaleres do SUS na região Norte desde Agosto/2007")
plt.legend(loc = 'upper right', bbox_to_anchor=(1.2, 1), frameon = True, shadow = True, ncol=1)
plt.show()
```

# DEsafio: Plotar as 5 UFs que mais gastaram com saúde

```
# Criação de uma coluna chamada "Total", que será utilizada para ordenar
usaveis["Total"] = usaveis.sum(axis="columns")

# Ordenação de acordo com as UFs que mais gastam
ordenado_por_total = usaveis.sort_values("Total",ascending=False)

# Remoção das colunas Total e Região para poder efetuar a conversão numérica da próxima célula
ordenado_por_total = ordenado_por_total.drop("Total", axis = "columns")
ordenado_por_total = ordenado_por_total.drop("Regiao", axis = "columns")
ordenado_por_total.head(30)
```

```
# Melhorar a leirura do grafico
ordenado_por_total_milhoes = ordenado_por_total/1000000 

# Repare que aqui o slice é antes de transpor o df, porque eu quero as 5 primeiras linhas dele, que são os Estados.
# Além disso, eu quero jogar fora aqueles meses que ocorreram aqueles picos que muito provavelmente foram causados por
# dados cujo atendimento era desconhecidos e foram computados no primeiro mês da série histórica, que foi Jan/2008
# Então, depois de transpôr o df, eu jogo fora as 6 primeiras linhas
ax = ordenado_por_total_milhoes[:5].T[6:].plot(figsize=(15,10), linewidth=4) 

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
ax.set_title("UFs com maiores gastos totais nos procedimentos hospitaleres do SUS desde Agosto/2007", fontsize=20)

ax.set_ylabel("GASTOS EM MILHÕES DE REAIS", fontsize=16)
# mudando o tamanho das fontes dos valores dos eixos
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

sns.set()

plt.legend(loc = 'upper right', bbox_to_anchor=(1.25, 1), frameon = True, shadow = True, ncol=1)
plt.show()
```

## Os valores do mês mais recente são flutuantes pois, além do mês em si mudar, o valor é atualizado semanalmente, o que leva a atualização dos valores.

```
# Vamos aproveitar para mudar o estilo do gráfico. O código abaixo mostra vários estilos disponíveis
print(plt.style.available)
```

```
ordenado_por_total_milhoes
```

```
plt.style.use("fivethirtyeight")
mes_mais_recente = ordenado_por_total_milhoes.columns[-1]
gastos_mes_mais_recente = ordenado_por_total_milhoes[mes_mais_recente]
gastos_mes_mais_recente.plot(kind="pie")
plt.show()
```

## No caso do gráfico acima, a vantagem é que ele está ordenado, por isso conseguimos identificar a ordem das UFs. Mas vamos ver como seria se ele não estivesse ordenado:

```
# O sample (que foi visto lá em cima) extrai uma amostra. O frac=1 é porque eu quero 100% da amostra, então basicamente ele vai reordenar aleatoriamente.
gastos_mes_mais_recente.sample(frac=1).plot(kind="pie")

# Percebe-se que tal gráfico é muito ruim para comparar
```

# Se a ideia é investigar o quanto uma UF gastou mais que outra, podemos fazer uma tabela simplesmente, dividindo pelo total da UF que queremos comparar.
# Por exemplo, vamos comparar quanto cada UF gastou em relação ao Piauí.
# Devemos sempre indicar as condições da tabela apresentada: valor total (e não por habitante) e mês mais recente cujo valor pode alterar

```
gastos_mes_mais_recente / gastos_mes_mais_recente.loc["22 Piauí"]
```

```
ax = (gastos_mes_mais_recente / gastos_mes_mais_recente.loc["22 Piauí"]).plot(kind="bar", figsize = (15,8))
ax.xaxis.grid(True)
ax.axhline(1, color='red', linewidth=1)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

ax.set_xlabel("")
ax.set_title("Gastos nos procedimentos hospitaleres do SUS de Jun/2020 relativos ao Piauí \n", fontsize=20, pad=15)

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0%}"))
plt.show()
```

## Repare que DF está acima da Paraíba, embora pela ordem seria esperado o contrário. É porque esse df está ordenado pela série histórica e não por esse mês. Para corrigir isso, é necessário reordenar:

```
valor_relativo_PI = (gastos_mes_mais_recente / gastos_mes_mais_recente.loc["22 Piauí"]).sort_values(ascending = False)
ax = (valor_relativo_PI).plot(kind="bar", figsize = (15,8))
ax.xaxis.grid(True)
ax.axhline(1, color='red', linewidth=1)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

ax.set_xlabel("")
ax.set_title("Gastos nos procedimentos hospitaleres do SUS de Jun/2020 relativos ao Piauí \n", fontsize=20, pad=15)

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0%}"))
plt.show()
```

## Agora, plotando em valores absolutos

```
ax = (gastos_mes_mais_recente.sort_values(ascending = False)).plot(kind="bar", figsize = (15,8))
ax.xaxis.grid(True)
ax.axhline(gastos_mes_mais_recente.loc["22 Piauí"], color='red', linewidth=1)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

ax.set_xlabel("")
ax.set_title("Gastos nos procedimentos hospitaleres do SUS de Jun/2020 \n", fontsize=20, pad=15)

ax.set_ylabel("R$ Milhões", fontsize=16)

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
plt.show()
```

# Desafio: Gasto por população

```
UF_pop = {
  "35 São Paulo": 44040000 ,
  "29 Bahia": 15130000,
  "22 Piauí": 3195000
}
```

```
mes_mais_recente = ordenado_por_total_milhoes.columns[-1]
gastos_mes_mais_recente = ordenado_por_total_milhoes[mes_mais_recente]

plt.style.use("fivethirtyeight")

# Voltar a considerar os milhões
gastos_mes_mais_recente = gastos_mes_mais_recente * 1000000


# Efetua o gasto/habitante
gasto_por_hab = (gastos_mes_mais_recente[["35 São Paulo","29 Bahia","22 Piauí"]]/[UF_pop["35 São Paulo"],UF_pop["29 Bahia"],UF_pop["22 Piauí"]]).sort_values(ascending=False)
gasto_por_hab.index = ["SP","PI","BA"]

# Parâmetro rot = 0 para rotacionar os labels do eixo x
ax = gasto_por_hab.plot(kind = "bar", figsize=(6,4),color=['olivedrab', 'sienna', 'green'],rot=0)
ax.set_title("Custo dos procedimentos hospitalares (SUS) \n por habitante em jullho/2020", fontsize=15, pad=15)
ax.set_xlabel("")
ax.set_ylabel("R$")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.show()
```
